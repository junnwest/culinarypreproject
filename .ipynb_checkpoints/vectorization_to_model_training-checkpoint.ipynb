# Splitting Dataset into Training and Testing Sets

This script splits the dataset (`train_final.json`) into:
- **80% Training Data** (`train_split.json`)
- **20% Testing Data** (`test_split.json`)

The files are saved in the `json_files` directory.

```python
import json
from sklearn.model_selection import train_test_split

file_path = 'json_files/train_final.json'
with open(file_path, 'r') as file:
    data = json.load(file)

train_split, test_split = train_test_split(data, test_size=0.2, random_state=42)

with open('json_files/train_split.json', 'w') as train_file:
    json.dump(train_split, train_file, indent=4)

with open('json_files/test_split.json', 'w') as test_file:
    json.dump(test_split, test_file, indent=4)

print(f"Training data saved to 'json_files/train_split.json' with {len(train_split)} recipes.")
print(f"Testing data saved to 'json_files/test_split.json' with {len(test_split)} recipes.")
```

# TF-IDF Vectorization of Ingredients

This script performs the following steps:
- **Loads Training and Testing Data**:
  - Training data: `train_split.json`
  - Testing data: `test_split.json`
- **Combines Ingredients**: Joins the list of ingredients into a single string separated by `"|||"`.
- **TF-IDF Vectorization**:
  - Transforms the ingredient strings into numerical feature vectors.
  - Uses a custom tokenizer for `"|||"`.

The vectorized data results in a feature matrix with:
- **Training Data Features**: `X_train_tfidf`
- **Testing Data Features**: `X_test_tfidf`

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import json

with open('json_files/train_split.json', 'r') as train_file:
    train_data = json.load(train_file)

with open('json_files/test_split.json', 'r') as test_file:
    test_data = json.load(test_file)

# Combine ingredients into a single string for each recipe, separated by "|||"
X_train = ["|||".join(recipe["ingredients"]) for recipe in train_data]
X_test = ["|||".join(recipe["ingredients"]) for recipe in test_data]

y_train = [recipe["cuisine"] for recipe in train_data]
y_test = [recipe["cuisine"] for recipe in test_data]

tfidf = TfidfVectorizer(tokenizer=lambda x: x.split("|||"), token_pattern=None)

X_train_tfidf = tfidf.fit_transform(X_train)

X_test_tfidf = tfidf.transform(X_test)

print(f"TF-IDF vectorized training data with {X_train_tfidf.shape[1]} features.")
print(f"TF-IDF vectorized test data with {X_test_tfidf.shape[1]} features.")
```

# PCA Reduction to 2 Dimensions for Visualization

This script reduces the high-dimensional TF-IDF vectors to **2 dimensions** using **TruncatedSVD**, allowing for visualization of the dataset in 2D space.

## Key Steps:
- **Dimensionality Reduction**:
  - Applied PCA to reduce TF-IDF features to 2 components.
  - Used `random_state=42` for reproducibility.
- **Training and Testing Data Shapes**:
  - **Training Data (2D)**: `X_train_pca_2`
  - **Testing Data (2D)**: `X_test_pca_2`

```python
from sklearn.decomposition import TruncatedSVD

pca_2 = TruncatedSVD(n_components=2, random_state=42)

X_train_pca_2 = pca_2.fit_transform(X_train_tfidf)
X_test_pca_2 = pca_2.transform(X_test_tfidf)

print(f"Reduced training data shape (2D): {X_train_pca_2.shape}")
print(f"Reduced testing data shape (2D): {X_test_pca_2.shape}")
```

```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

pca_df = pd.DataFrame(X_train_pca_2, columns=["PC1", "PC2"])
pca_df["cuisine"] = y_train  # Add cuisine labels

cuisine_centroids = pca_df.groupby("cuisine")[["PC1", "PC2"]].mean()

n_clusters = 3
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
cluster_labels = kmeans.fit_predict(cuisine_centroids)

cuisine_centroids["Cluster"] = cluster_labels

plt.figure(figsize=(12, 8))
scatter = plt.scatter(
    cuisine_centroids["PC1"],
    cuisine_centroids["PC2"],
    c=cuisine_centroids["Cluster"],
    cmap="viridis",
    s=200,
    alpha=0.8,
)

for cuisine, row in cuisine_centroids.iterrows():
    plt.text(row["PC1"], row["PC2"], f"{cuisine}", fontsize=9, ha='center', alpha=0.9)

plt.title("Cuisine Centroids in 2D PCA Space with K-Means Clusters")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.colorbar(scatter, label="Cluster")
plt.show()
```

# PCA for Preprocessing: Finding Optimal Number of Components

This script determines the optimal number of components for PCA to retain a desired amount of explained variance.

## Key Steps:
1. **Fit PCA on the Training Data**:
   - Applied PCA to the full TF-IDF feature set.
   - No initial limit on the number of components.

2. **Calculate Cumulative Explained Variance**:
   - Plotted the cumulative explained variance ratio against the number of components.
   - Marked thresholds for **90%** and **95%** explained variance.

3. **Determine Optimal Number of Components**:
   - **90% Variance**: Number of components required to retain ~90% of the variance.
   - **95% Variance**: Number of components required to retain ~95% of the variance.

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import TruncatedSVD

pca_full = TruncatedSVD(n_components=min(X_train_tfidf.shape[0], X_train_tfidf.shape[1]), random_state=42)

pca_full.fit(X_train_tfidf)

cumulative_variance = pca_full.explained_variance_ratio_.cumsum()

plt.figure(figsize=(8, 5))
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o', linestyle='--')
plt.axhline(y=0.9, color='r', linestyle='--', label="90% Variance")
plt.axhline(y=0.95, color='g', linestyle='--', label="95% Variance")
plt.title('Cumulative Explained Variance vs. Number of Components')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.legend()
plt.grid()
plt.show()

n_components_90 = next(i for i, ratio in enumerate(cumulative_variance) if ratio >= 0.9) + 1
print(f"Number of components to retain 90% variance: {n_components_90}")

n_components_95 = next(i for i, ratio in enumerate(cumulative_variance) if ratio >= 0.95) + 1
print(f"Number of components to retain 95% variance: {n_components_95}")
```

```python
from sklearn.decomposition import TruncatedSVD

n_components = 1468  # Retain 90% variance
pca_90 = TruncatedSVD(n_components=n_components, random_state=42)

# Transform the training and testing datasets
X_train_pca_90 = pca_90.fit_transform(X_train_tfidf)
X_test_pca_90 = pca_90.transform(X_test_tfidf)

print(f"Reduced training data shape: {X_train_pca_90.shape}")
print(f"Reduced testing data shape: {X_test_pca_90.shape}")
```

# Model Training and Comparison

This script trains and evaluates multiple machine learning models on the PCA-reduced dataset and compares their accuracy.

## Models Evaluated
- **K-Nearest Neighbors (KNN)**
- **Naive Bayes**
- **Logistic Regression**
- **Random Forest**
- **Decision Tree**
- **Support Vector Machine (SVM)**
- **Stochastic Gradient Descent (SGD)**

## Workflow
1. **Model Training**:
   - Each model is trained using the PCA-reduced training data (`X_train_pca_90` and `y_train`).
2. **Prediction**:
   - Predictions are made on the PCA-reduced test data (`X_test_pca_90`).
3. **Evaluation**:
   - Accuracy scores are calculated using `accuracy_score`.
4. **Comparison**:
   - Results are stored in a DataFrame and sorted by accuracy for comparison.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
import pandas as pd

models = {
    "KNN": KNeighborsClassifier(n_neighbors=5),
    "Naive Bayes": MultinomialNB(),
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "SVM": SVC(random_state=42),
    "SGD": SGDClassifier(max_iter=1000, random_state=42)
}

results = []

for model_name, model in models.items():
    print(f"Training {model_name}...")
    try:
        model.fit(X_train_pca_90, y_train)
        
        y_pred = model.predict(X_test_pca_90)
        
        accuracy = accuracy_score(y_test, y_pred)
        results.append({"Model": model_name, "Accuracy": accuracy})
        
        print(f"{model_name} Accuracy: {accuracy:.4f}")
    except Exception as e:
        print(f"{model_name} failed: {e}")

results_df = pd.DataFrame(results).sort_values(by="Accuracy", ascending=False)

print("\nModel Comparison:")
print(results_df)
```
